```python
import pandas as pd
import numpy as np
import os

path = 'D:\\maran10\\Documents\\日常工作\\B.Q2业绩核算【已完成存档】\\2.商品池数据源\\Q2商用商品池数据源\\'
path_new = 'D:\\maran10\\Documents\\日常工作\\6.数据复盘\\4.其他数据源\\2020Q2商用订单明细\\'
# path_goldeye = path+'1.2020H1黄金眼\\原数据\\'
# path_goldeye_new = path+'1.2020H1黄金眼\\加上行业后\\'
```

```python
count=0
ori_Q2 = os.listdir(path)

for f in ori_Q2:
    if not ".xlsx" in f:
        continue
    ori = pd.read_excel(path+f)
    print(count,len(ori))
    
    if count == 0:
        ori_total = pd.DataFrame(columns = list(ori))
    else:
        ori_total = pd.concat([ori,ori_total])
    count = count+1

for name, group in ori_total.groupby(['区域','一级部门']):
    print(name)
    group.to_excel(path_new+str(name)+"Q1商用商品池订单明细.xlsx",encoding = 'utf-8-sig',index = False)
```
